```python
#Import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import re

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create

#read the data
df = pd.read_csv(r'C:\Users\Dinh Tien Vu\Downloads\Data Base\movies.csv')
```

```python
#Look at the data
df.head()
```

```python
# check missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# delete the rows, which is missing values
df = df.dropna()
```

```python
#Check missing data again
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#data types for our cols
df.dtypes
```

```python
#change data type of cols
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#create correct year col
def extract_year(date_str):
    if isinstance(date_str, str):  # Kiểm tra xem giá trị có phải là chuỗi không
        # Tìm năm trong chuỗi (4 chữ số)
        year_match = re.search(r'(\d{4})', date_str)
        return year_match.group(1) if year_match else None  # Trả về năm hoặc None nếu không tìm thấy
    return None  # Nếu giá trị không phải chuỗi, trả về None

# Áp dụng hàm trích xuất năm và tạo một cột mới 'correctyear'
df['correctyear'] = df['released'].apply(extract_year)
df
```

```python
#order by gross
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
#Drop any duplicates
df['company'].drop_duplicates().sort_values(ascending = False).head()
```

```python
# Looking at correlation
df_numeric = df.select_dtypes(include=['number'])
df_numeric.corr()
```

```python
# -> high correlation between budget and gross
correlation_matrix = df_numeric.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
